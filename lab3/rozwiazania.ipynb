# Zadanie 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open("images/chill_guy.webp")
image_np = np.array(image)

mask = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

filtered_image = cv2.filter2D(image_np, -1, mask)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(image_np, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Po filtrze górnoprzepustowym")
plt.imshow(filtered_image, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```

# Zadanie 2

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open("images/chill_guy.webp")
img = np.array(image) / 255.0 

M = np.array([
    [0.393, 0.769, 0.189],
    [0.349, 0.689, 0.168],
    [0.272, 0.534, 0.131]
])

img_new = img @ M.T
img_new = np.clip(img_new, 0.0, 1.0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Po przekształceniu kolorów")
plt.imshow(img_new)
plt.axis("off")

plt.tight_layout()
plt.show()
```

# Zadanie 3

```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

img = np.array(Image.open("images/chill_guy.webp"))

M = np.array([
    [0.229,  0.587,  0.114],
    [0.500, -0.418, -0.082],
    [-0.168, -0.331,  0.500]
])
bias = np.array([0.0, 128.0, 128.0])

rgb = img.astype(np.float32)
ycrcb = rgb @ M.T + bias
ycrcb = np.clip(ycrcb, 0, 255).astype(np.uint8)

Y, Cr, Cb = ycrcb[:, :, 0], ycrcb[:, :, 1], ycrcb[:, :, 2]

image_rgb_restored = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2RGB)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(img)
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title("Y (jasność)")
plt.imshow(Y, cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title("Cr (czerwona składowa)")
plt.imshow(Cr, cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title("Cb (niebieska składowa)")
plt.imshow(Cb, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
