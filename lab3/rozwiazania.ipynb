# Zadanie 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open("images/chill_guy.webp")
image_np = np.array(image)

mask = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

filtered_image = cv2.filter2D(image_np, -1, mask)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(image_np, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Po filtrze górnoprzepustowym")
plt.imshow(filtered_image, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```

# Zadanie 2

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open("images/chill_guy.webp")
img = np.array(image) / 255.0

R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]

R_new = 0.393 * R + 0.769 * G + 0.189 * B
G_new = 0.349 * R + 0.689 * G + 0.168 * B
B_new = 0.272 * R + 0.534 * G + 0.131 * B

img_new = np.stack([R_new, G_new, B_new], axis=2)
img_new = np.clip(img_new, 0.0, 1.0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(img)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Po przekształceniu kolorów")
plt.imshow(img_new)
plt.axis("off")

plt.tight_layout()
plt.show()

```

# Zadanie 3

```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

img = np.array(Image.open("images/chill_guy.webp"))

R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]

Y  = 0 + (0.229 * R + 0.587 * G + 0.114 * B)
Cr = 128.0 + (0.500 * R - 0.418 * G - 0.082 * B)
Cb = 128.0 + (-0.168 * R - 0.331 * G + 0.500 * B)

Y  = np.clip(Y, 0, 255).astype(np.uint8)
Cr = np.clip(Cr, 0, 255).astype(np.uint8)
Cb = np.clip(Cb, 0, 255).astype(np.uint8)

ycrcb = np.stack([Y, Cr, Cb], axis=2)

image_rgb_restored = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2RGB)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(img)
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title("Y")
plt.imshow(Y, cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title("Cr (składowa czerwona)")
plt.imshow(Cr, cmap="gray")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title("Cb (składowa niebieska)")
plt.imshow(Cb, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
